### Текст на графіках

xlabel (або ylabel) — рядок тексту підпису.
А також параметри конструктора класу matplotlib.text.Text:

`fontsize` або `size` — розмір шрифту, число або значення зі списку `xx-small`, `x-small`, `small`, `medium`, `large`, `x-large`, `xx-large`

`fontstyle` — стиль шрифту значення зі списку normal, italic, oblique
`fontweight` — товщина шрифту може визначатися або числом в діапазоні від 0 до 1000, або значенням зі списку `ultralight`, `light`, `normal`, `regular`, `book`, `medium`, `roman`, `semibold`, `demibold`, `demi`, `bold`, `heavy`, `extra bold`, `black`
`color` — значення кольору для тексту підпису. Можливості вибору кольору занадто великі і подивитися варіанти можна за цим посиланням (https://matplotlib.org/stable/gallery/color/named_colors.html), де вони теж наведені не повністю.

```python
date = pd.date_range(start='2021-09-01', freq='D', periods=8)
plt.figure(figsize=(5,6))
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.xticks(fontsize=5)
plt.yticks(fontsize=10)
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Температура в м. Полтава', fontsize=15, fontweight=100)
plt.text(date[0], 20, 'Осінь досить тепла', color="blue")
plt.legend()
plt.show()
```

```python
# Компонування графіків

date = pd.date_range(start='2021-09-01', freq='D', periods=8)
fig, axs = plt.subplots()
axs.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
axs.plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Температура в м. Полтава', fontsize=15)
plt.text(date[0], 15, 'Осінь досить тепла', color="blue")
plt.legend()
plt.show()
```

```python
# кількість рядків та стовпців

date = pd.date_range(start='2021-09-01', freq='D', periods=8)
#plt.figure(figsize=(30,50))
fig, axs = plt.subplots(2, 1)
fig.tight_layout(pad=5.0)

axs[0].plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
axs[1].plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature')

axs[0].set_title('Денна', fontsize=10)
axs[1].set_title('Нічна', fontsize=10)

fig.suptitle('Температура в м. Полтава', fontsize=15)

plt.show()
```

```python
# Див: стиль лінії графіка, задається через параметр linestyle

date = pd.date_range(start="2021-09-01", freq="D", periods=8)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label="day temperature",
    linestyle="--",
    color="#FF5733",
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label="night temperature",
    linestyle=":",
    color="#061358",
)
plt.xlabel("Дата", fontsize="small", color="midnightblue")
plt.ylabel("Температура", fontsize="small", color="midnightblue")
plt.title("Температура в м. Полтава", fontsize=15)
plt.legend()
plt.show()
```

```python
# плюс точки, товщина лінії і сітка
date = pd.date_range(start="2021-09-01", freq="D", periods=8)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label="day temperature",
    linestyle="--",
    color="#FF5733",
    linewidth=2,
    marker="D",
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label="night temperature",
    linestyle=":",
    color="#061358",
    linewidth=2,
    marker="^",
)
plt.ylim(0, 25)
plt.xlabel("Дата", fontsize="small", color="midnightblue")
plt.ylabel("Температура", fontsize="small", color="midnightblue")
plt.title("Температура в м. Полтава", fontsize=15)
plt.legend()
plt.grid(color='r', linestyle='-', linewidth=2)
plt.show()
```

```python
# Стовпчикові діаграми
# see: parameters

plt.bar(
    ["США", "Китай", "Японія", "Велика Британія"],
    [39, 38, 27, 22],
    color=["b", "r", "y", "g"]#, yerr=0.6#, hatch='+'
)

plt.xlabel("Країна", fontsize="small", color="midnightblue")
plt.ylabel("Кількість", fontsize="small", color="midnightblue")
plt.title("Золоті медалі: Літня олімпіада Токіо 2020", fontsize=15)
plt.show()
```

```python
# Діаграми в полярних координатах
theta = np.linspace(0, 2.0 * np.pi, 1000)

r = np.sin(6 * theta)
plt.polar(theta, r)
plt.show()
```

```python
# Тривимірні графіки
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

theta_max = 8 * np.pi
n = 1000
theta = np.linspace(0, theta_max, n)
x = theta
z = np.sin(theta)
y = np.cos(theta)
ax.plot(x, y, z, "g")

plt.show()
```

```python
# Діаграма розсіювання

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

x = [5, 10, 15, 20]
z = [10, 0, 5, 15]
y = [0, 10, 5, 25]
s = [150, 130, 30, 160]
ax.scatter(x, y, z, s=s)

plt.show()
```

```python
# Каркасна поверхня
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_wireframe(x, y, z)

plt.show()
```

```python
# Поверхня
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_surface(x, y, z)

plt.show()
```

# Seaborn

```python
import seaborn as sns
```

```python
# data from https://github.com/mwaskom/seaborn-data/blob/master/mpg.csv
# Набір mpg містить інформацію про характеристики ряду автомобілів
data = sns.load_dataset("mpg")
```

```python
data
```

```python
# стиль оформлення графіків
# глобальне налаштування
sns.set_style("whitegrid")
```

```python
date = pd.date_range(start="2021-09-01", freq="D", periods=8)
day = [23, 17, 17, 16, 15, 14, 17, 20]
night = [19, 11, 16, 11, 10, 10, 11, 16]
df = pd.DataFrame({'date':date, 'day_temperature': day, 'night_temperature': night})
df.head()
```

```python
# Функція lineplot
sns.lineplot(data=df)
```

```python
# графік залежності нічної температури від денної
# Світліша область навколо лінії – це 95% довірчий інтервал
sns.lineplot(x="night_temperature", y="day_temperature", data=df)
```

```python
#за замовчуванням seaborn здійснює сортування набору даних
sns.lineplot(x="night_temperature", y="day_temperature", data=df, sort=False)
```

```python
#графік залежності потужності автомобіля від року випуску окремо для кожного регіону
sns.set_style("darkgrid")
data = sns.load_dataset("mpg")
sns.lineplot(x='model_year', y='horsepower', hue="origin", data=data)
```

```python
#точковий графік залежності дальності пробігу (displacement) автомобіля від витрат палива (mpg)
data = sns.load_dataset("mpg")
sns.scatterplot(x='mpg', y='displacement', data=data)
```

```python
# можемо виділити кольором регіон виробника
sns.scatterplot(x='mpg', y='displacement', hue='origin', data=data)
```

```python
#ця функція надає можливість налаштовувати не лише зовнішній вигляд графіка, 
# але й фігуру - підкладку, на якій розміщені всі графічні компоненти

sns.relplot(x='mpg', y='displacement', kind='scatter', data=data)
```

```python
# діаграму можна розмістити на трьох окремих полях, як роздільний параметр будемо використовувати регіон (origin)
sns.relplot(x='mpg', y='displacement', kind='scatter', hue='origin', col='origin', data=data)
```

```python
# Візуалізація категоріальних даних
# вони приймають значення з невпорядкованої множини. Це ознаки кольору, форми тощо
# до них не можна застосувати операцію порівняння
sns.stripplot(x='origin', y='mpg', data=data)
```

```python
#виділимо кольором кількість циліндрів у двигуна
sns.stripplot(x='origin', y='mpg', hue='cylinders', data=data)
```

```python
#точки, що відображаються на діаграмі, не перекриваються, це дозволяє робити висновки про переважання тих або інших значень в наборах даних 
sns.swarmplot(x='origin', y='mpg', data=data)
```

```python
# Візуалізації розподілу категоріальних даних

# будує діаграму типу "скринька з вусами", на ній відображаються медіанне значення, квантилі та викиди
sns.boxplot(x="origin", y="mpg", hue="cylinders", data=data)
```

```python
# density
sns.violinplot(x="origin", y="mpg", data=data)
```

```python
#будує діаграму з прямокутників, добре підходить для візуалізації великих наборів даних.
sns.boxenplot(x="origin", y="mpg", data=data)
```

```python
# Візуалізація оцінок категоріальних даних

# оцінку будь-якого набору даних як точку на полі графіка та довірчий інтервал у вигляді лінії, центр якої лежить на зазначеній точці
sns.pointplot(x="origin", y="mpg", data=data)
```

```python
#будує стовпчасту діаграму: висота бару (стовпця) визначає чисельне значення оцінки ознаки (математичне очікування), лінія, що перетинає верхню межу бару - довірчий інтервал.
sns.barplot(x="origin", y="mpg", data=data)
```

```python
# визначає кількість елементів з набору даних, які належать до тієї або іншої категорії, і відображає отримане значення у вигляді стовпчастої діаграми.
sns.countplot(x="cylinders", data=data)
```

```python
#відображає набір даних та лінію регресії, побудовану за ними.
#лінію регресії залежності відстані, яку може проїхати автомобіль, від його потужності.
sns.lmplot(x="horsepower", y="displacement", data=data)
```

```python
#діаграмa з медіанною оцінкою та необхідні параметри функції при цьому
from numpy import median
sns.lmplot(x="horsepower", y="displacement", data=data, x_estimator=median)
```

```python
#відхилення елементів вихідного набору даних від регресійної моделі, побудованої за ними, у вигляді діаграми розсіювання. 
# Кожна точка такої діаграми – це різниця між значенням елемента вихідного набору та значенням, яке видасть модель регресії у цій точці

sns.residplot(x="horsepower", y="displacement", data=data)
```

```python
#Функція повний аналог функції regplot з можливістю управляти компонуванням полів з графіками на підкладці як для функції relplot.

sns.lmplot(x="horsepower", y="displacement", hue="origin", col="origin", data=data)
```
